## Import and clean raw data

```python
import pandas as pd
import glob
```

```python
# Get all our files and create a blank list to hold them.
all_files = glob.glob("/content/drive/MyDrive/Data/Raw/LFS/*.csv")
li = []

# Loop through all the files, add a new column so we can keep track of what PHU they're from, and add them to our list.
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

# Concatenate all the dataframes into one big one, which will be our raw data.
raw = pd.concat(li, axis=0, ignore_index=True)
raw = raw.set_index('REC_NUM')

display(raw)
```

# Overtime

```python
overtime = raw.copy()
overtime["Number of overtime or extra hours worked"] = overtime["Number of overtime or extra hours worked"].fillna(0)
```

## Average overtime for nurses

```python
overtime_nurses = overtime[overtime["Occupation at main job - Detail"] == "Professional occupations in nursing"]
overtime_nurses = overtime_nurses.groupby(["Province", "Survey date"]).mean().reset_index()

overtime_nurses_pivot = pd.pivot(overtime_nurses, index=["Province"], columns="Survey date", values="Number of overtime or extra hours worked")

display(overtime_nurses_pivot)
```

## Overtime for various professions

```python
# overtime_october = overtime_october.groupby(["Occupation at main job - Detail"]).mean().reset_index()
overtime_all = overtime.groupby(["Occupation at main job - Detail", "Survey date"]).mean().reset_index()

overtime_all_pivot = pd.pivot(overtime_all, index=["Occupation at main job - Detail"], columns="Survey date", values="Number of overtime or extra hours worked")

overtime_all_pivot = overtime_all_pivot.sort_values("2021-10-01", ascending=False)

display(overtime_all_pivot)
```
