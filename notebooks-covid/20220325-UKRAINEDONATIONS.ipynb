```python
import pandas as pd
```

```python
raw = pd.read_csv('../raw/RAW 2022 UKRAINE FUNDING.csv')

raw.head()
```

```python
all_donations = (raw
                 .pivot_table(index="Private sector donor (company or foundation)", aggfunc="sum")
                 )

all_donations["total_donations"] = (((all_donations["Est USD value cash donation"] + all_donations["Est USD value in-kind donation"]) * 1.25)
                                    .astype(int)
                                    )

all_donations = all_donations[["total_donations"]].sort_values("total_donations", ascending=False)

all_donations.head()
```

Let's calculate the total quickly.

```python
all_donations.sum()
```

### What charities are the biggest recipients of this funding?

```python
recipients = (raw
                 .pivot_table(index="Recipient", aggfunc="sum")
                 ) * 1.25

recipients["total_donations"] = (((recipients["Est USD value cash donation"] + recipients["Est USD value in-kind donation"]) * 1.25)
                                    .astype(int)
                                    )

recipients = recipients[["total_donations"]].sort_values("total_donations", ascending=False)

recipients.head()
```
