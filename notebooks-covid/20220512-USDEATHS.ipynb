```python
import pandas as pd
```

```python
raw = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')

raw
```

```python
us = raw.loc[raw["location"] == "United States", ["date", "new_deaths", "total_deaths"]].set_index("date").dropna(how="all")
us["new_deaths"] = us["new_deaths"].rolling(7).mean()

us.loc[us["total_deaths"].between(0, 250000), "0 to 25K"] = us["new_deaths"]
us.loc[us["total_deaths"].between(249999, 500000), "25K to 50K"] = us["new_deaths"]
us.loc[us["total_deaths"].between(499999, 750000), "50K to 75K"] = us["new_deaths"]
us.loc[us["total_deaths"].between(749999, 1000000), "75K to 1M"] = us["new_deaths"]

us
```

```python
million_countries = raw[raw["total_deaths"] >= 1000000]
million_countries["location"].unique()
```
