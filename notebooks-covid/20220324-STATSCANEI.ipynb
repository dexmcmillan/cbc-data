# How has Employment Insurance usage changed over time?

*March 24, 2022*

Let's take a look at some data released by StatsCan today on uptake of EI in Canada.

```python
import pandas as pd

raw = (pd
       .read_csv('../raw/RAW 2022 EI.csv', skip_blank_lines=True)
       .dropna(how="all")
       )

display(raw.head())
```

Finally, we'll do a reshape so we can graph this easily.

```python
canada = (raw
          .loc[(raw["GEO"] == "Canada"), ["Occupations", "VALUE", "REF_DATE"]]
          .pivot(columns="Occupations", index="REF_DATE", values="VALUE")
          .astype(int)
          )

canada.columns = (canada.columns
                  .str.replace("[\[0-9\]]", "", regex=True)
                  .str.strip()
                  )

canada.head()
```
