```python
import pandas as pd
```

```python
raw = pd.read_csv('/content/drive/MyDrive/Data/raw/COVID-19_Reported_Patient_Impact_and_Hospital_Capacity_by_State_Timeseries.csv')

display(raw)
```

```python
nov22 = raw[raw["date"].str.contains('/22')]
nov22 = nov22[nov22["state"] != "AS"]
nov22 = nov22[["date", "state", "hospital_onset_covid"]]

dates = nov22["date"].unique()

ranked = []

for date in dates:
  top10 = nov22[nov22["date"] == date].sort_values('hospital_onset_covid', ascending=False)
  top10["rank"] = range(1, len(top10)+1)
  ranked.append(top10)

all_ranked = pd.concat(ranked)
display(all_ranked)
```

```python
all_ranked = all_ranked[all_ranked["date"] >= "2021/01/01"]

pivot = pd.pivot(all_ranked, columns="date", index="state", values="rank")

display(pivot)
```
