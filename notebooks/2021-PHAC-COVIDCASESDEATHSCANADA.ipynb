# Health Canada COVID-19 Data Analysis

A collection of data analysis of Health Canada's COVID data. Raw data can be found [here](https://health-infobase.canada.ca/src/data/covidLive/covid19-download.csv).

## IMPORT - Modules and raw data

```python
import pandas as pd

# Read in the data we're using from Health Canada.
raw = pd.read_csv('https://health-infobase.canada.ca/src/data/covidLive/covid19-download.csv')

display(raw)
```

```python
onque = raw[raw["prname"] == "Ontario"]
onque = onque[onque["date"] >= "2021-12-01"]

onque = onque[["numdeaths", "num"]]

display(onque)
```
