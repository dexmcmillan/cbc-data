# COVID-19 attitudies survey

```python
import pandas as pd
```

```python
raw = pd.read_csv("../raw/RAW 2021 STATSCAN COVID SURVEY.csv")
```

```python
working = (raw[(raw["Characteristics"] == "Percentage of persons")
               & (raw["Age group"] == "Total, 12 years and over")
               & (raw["Gender"] == "Total, gender of person")]
           .pivot(index="Indicators", columns="Collection period", values="VALUE")
           )

working["Change since Feb. 2021"] =  working["September to mid-November 2021"] - working["January to February 2021"]
working["direction"] = working["Change since Feb. 2021"]
working["direction"][working["Change since Feb. 2021"] < 0] = "Decrease"
working["direction"][working["Change since Feb. 2021"] > 0] = "Increase"

display(working)
working.to_clipboard()
```
