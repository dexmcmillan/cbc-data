```python
import pandas as pd

pd.options.display.float_format = '{:,.2f}'.format
```

```python
raw = pd.read_csv("../raw/RAW 2022 UKRAINE DONATIONS.csv")

country_codes = pd.read_csv("../raw/RAW 2022 DATAWRAPPER COUNTRY CODES.csv")

raw["total_donations"] = (raw["Est USD value cash donation"] + raw["Est USD value in-kind donation"]) * 1.25

```

```python
companies = (raw
             .pivot_table(index=["Private sector donor (company or foundation)", "Country of Registration"], values="total_donations", aggfunc='sum')
             .sort_values("total_donations", ascending=False)
             .astype(float)
             .reset_index()
             .merge(country_codes, how="left", left_on="Country of Registration", right_on="Country")
             .dropna(how="any")
             )

companies["Private sector donor (company or foundation)"] = companies["Code"] + " " + companies["Private sector donor (company or foundation)"]

companies = (companies.loc[companies["total_donations"] > 0, ["Private sector donor (company or foundation)", "Country of Registration", "total_donations"]]
            .set_index("Private sector donor (company or foundation)")
            )
companies.to_clipboard()
companies
```

```python
countries = (raw
             .pivot_table(index="Country of Registration", values="total_donations", aggfunc='sum')
             .sort_values("total_donations", ascending=False)
             .astype(float)
             )

countries
```

```python
canada = raw.loc[raw["Country of Registration"] == "Canada", :]
canada = (canada
            .loc[:, ["Private sector donor (company or foundation)", "total_donations", "One-sentence description for dataviz"]]
            .groupby(["Private sector donor (company or foundation)"]).sum()
            .sort_values("total_donations", ascending=False)
            )

display(canada)
canada.to_clipboard()
```

```python
cayman = raw.loc[raw["Country of Registration"] == "Cayman Islands", :]

cayman = (cayman
            .loc[:, ["Private sector donor (company or foundation)", "total_donations", "One-sentence description for dataviz"]]
            .groupby(["Private sector donor (company or foundation)"]).sum()
            .sort_values("total_donations", ascending=False)
            )

cayman
```
