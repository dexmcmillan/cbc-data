```python
import pandas as pd

raw = pd.read_csv("../raw/RAW 2022 UKRAINEIDP.csv")

display(raw.head())
```

```python
raw = raw[raw.columns[::-1]]

display(raw)
```

```python
vars = [
        "IDP estimation",
        "IDP location Cluster",
        "Y Latitude",
        "X Longitude",
        "admin1Name_eng",
        "admin1ClassType",
        "admin2Name_eng",
        "admin2ClassType",
        "admin3Name_eng",
        "admin4Name_eng",
        "admin4ClassType",
            ]

data = (raw.loc[:, vars]
        )

display(data.head())
```

```python
data[data["IDP location Cluster"] == 3]
```

```python
melt = (pd
        .melt(data, id_vars=["admin1Name_eng",  "admin2Name_eng", "admin4Name_eng"], value_vars=["IDP estimation"])
        )

display(melt.head())
```

```python
pivot = melt.pivot_table(columns="variable", index="admin1Name_eng", aggfunc="sum")

pivot.index = [
    "Cherkasy Oblast",
    "Chernivtsi Oblast",
    "Dnipropetrovsk Oblast",
    "Ivano-Frankivsk Oblast",
    "Khmelnytskyi Oblast",
    "Kirovohrad Oblast",
    "Lviv Oblast",
    "Mykolaiv Oblast",
    "Odesa Oblast",
    "Poltava Oblast",
    "Rivne Oblast",
    "Ternopil Oblast",
    "Vinnytsia Oblast",
    "Volyn Oblast",
    "Zakarpattia Oblast",
    "Zaporizhzhia Oblast"
]

missing = [
    "Chernihiv Oblast",
    "Donetsk Oblast",
    "Kharkiv Oblast",
    "Kherson Oblast",
    "Kyiv Oblast",
    "Luhansk Oblast",
    "Sumy Oblast",
    "Zhytomyr Oblast"
]

display(pivot)
pivot.to_clipboard()
```
