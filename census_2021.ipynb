```python
import pandas as pd
```

## CMA Population Changes from 2016 to 2021

```python
raw = pd.read_csv("raw/census2021.csv", encoding = 'unicode_escape', engine ='python')

display(raw.head(5))
```

```python
population_raw = raw[raw["CHARACTERISTIC_NAME"].isin(["Population percentage change, 2016 to 2021"])]

display(population_raw)
```

```python
population_pivot = pd.pivot(population_raw, index=["ALT_GEO_CODE", "GEO_NAME"], columns="CHARACTERISTIC_NAME", values="C1_COUNT_TOTAL").reset_index().set_index("ALT_GEO_CODE")
display(population_pivot)

population_pivot.to_csv("exports/populations2021.csv")
```

```python
cmas = population_raw[["ALT_GEO_CODE", "GEO_NAME"]].drop_duplicates(subset="ALT_GEO_CODE")
code = cmas["ALT_GEO_CODE"].astype(str).str[-3:].astype(int)
cmas["ALT_GEO_CODE"] = code


display(cmas)
cmas.to_csv("exports/cmas_list.csv")
```

## Census subdivisions - Change from 2011 to 2016

```python
raw_csd = pd.read_csv("raw/2021_census_csd.csv", encoding = 'unicode_escape', engine ='python')

display(raw_csd)
```

```python
csd = raw_csd[raw_csd["GEO_LEVEL"] == "Census subdivision"]
csd = csd[csd["CHARACTERISTIC_NAME"].isin(["Population percentage change, 2016 to 2021", "Population, 2021"])]

csd_pivot = pd.pivot(csd, index=["DGUID", "GEO_NAME", "ALT_GEO_CODE"], columns="CHARACTERISTIC_NAME", values="C1_COUNT_TOTAL").reset_index()

display(csd_pivot)
csd_pivot.to_csv("exports/cts_formapping.csv")
```

```python
csd_pivot = pd.pivot(csd, index=["DGUID", "GEO_NAME"], columns="CHARACTERISTIC_NAME", values="C1_COUNT_TOTAL").reset_index()
name = csd_pivot["GEO_NAME"].str.split(", ", expand=True)[0]
csd_pivot["TYPE"] = csd_pivot["GEO_NAME"].str.split(", ", expand=True)[1]
csd_pivot["GEO_NAME"] = name

display(csd_pivot)
csd_pivot.to_csv("exports/census2021.csv")
csd_pivot.to_clipboard()
```

## Census divisions

```python
raw_csd = pd.read_csv("raw/2021_census_csd.csv", encoding = 'unicode_escape', engine ='python')
sd = raw_csd[raw_csd["GEO_LEVEL"] == "Census division"]
sd = sd[sd["CHARACTERISTIC_NAME"] == "Population percentage change, 2016 to 2021"]

display(sd)
```

```python
sd_table = sd[["ALT_GEO_CODE", "GEO_NAME", "C1_COUNT_TOTAL"]].set_index("ALT_GEO_CODE")

sd_list = sd_table.drop_duplicates()
sd_list.to_clipboard()
```

```python
sd_clean = sd_table.copy()
sd_clean["GEO_NAME"] = sd_clean["GEO_NAME"].astype(str).str.replace(pat=", Region (REG)", repl="",regex=False)
sd_clean["GEO_NAME"] = sd_clean["GEO_NAME"].astype(str).str.replace(pat=", Regional district (RD)", repl="",regex=False)
sd_clean["GEO_NAME"] = sd_clean["GEO_NAME"].astype(str).str.replace(pat=", District (DIS)", repl="",regex=False)
sd_clean["GEO_NAME"] = sd_clean["GEO_NAME"].astype(str).str.replace(pat=", County (CTY)", repl="",regex=False)
sd_clean["GEO_NAME"] = sd_clean["GEO_NAME"].astype(str).str.split(pat=" / ", expand=True)[0]

display(sd_clean)
sd_clean.to_csv("exports/2021divisions.csv")
```
