```python
import pandas as pd
import re
```

```python
filepaths = [
    "./data/2020-Offence-Based-Criminal.xlsx",
    "./data/2021-Offence-Based-Criminal.xlsx"
    ]

lis = []

for filepath in filepaths:
    
    file = pd.ExcelFile(filepath)
    
    for sheet in file.sheet_names:
        print(sheet)
        df = pd.read_excel(filepath, sheet_name=sheet, skiprows=6, header=[0,1], skipfooter=12)
        df["REGION"] = sheet
        df["YEAR"] = re.search('[0-9]{4}', filepath).group(0)
        lis.append(df)

raw = pd.concat(lis)

raw
```

```python
raw["REGION"].unique()
```

```python
data = raw.copy()
data.columns =  [' '.join(col).strip() for col in data.columns.values]
data.columns = (data.columns
                .str.replace(" Unnamed: [0-9]+_level_1", "", regex=True)
                .str.replace("\n", " ", regex=True)
                .str.upper()
                .str.strip()
                )
data = data.drop(columns=["CASES DISPOSED BEFORE TRIAL DATE TOTAL", "CASES DISPOSED AT TRIAL WITHOUT TRIAL TOTAL"])


data["OFFENCE GROUP"] = data["OFFENCE GROUP"].ffill()
data = data.loc[(data["OFFENCE TYPE"] != "TOTAL") & (data["OFFENCE GROUP"] != "TOTAL CASES")]

data
```

```python
values = data.iloc[:,0:15]

db = data.melt(id_vars=["YEAR", "REGION", "OFFENCE GROUP", "OFFENCE TYPE"], value_vars=values)

db
```

```python
lis = []

for year in ["2020", "2021"]:
    stay_rates = (db
            .loc[(db["YEAR"] == year) & (db["variable"].isin(["CASES DISPOSED BEFORE TRIAL DATE WITHDRAWN/ STAYED BEFORE TRIAL", "CASES DISPOSED"]))]
            .pivot(columns=["variable"], index=["REGION", "OFFENCE TYPE"], values="value")
            )

    stay_rates["withdrawn_%"] = round(stay_rates["CASES DISPOSED BEFORE TRIAL DATE WITHDRAWN/ STAYED BEFORE TRIAL"] / stay_rates["CASES DISPOSED"] * 100, 2)

    stay_rates.loc[stay_rates["CASES DISPOSED"] <= 30, "withdrawn_%"] = pd.NA

    stay_rates = stay_rates.loc[:, ["withdrawn_%"]].reset_index().pivot(columns="OFFENCE TYPE", index="REGION", values="withdrawn_%")
    lis.append(stay_rates)

diff = lis[1].sub(lis[0])

diff
```

```python
diff.plot()
```

## Stay/withdrawal rates

```python
drugs_2020 = (db
         .loc[(db["OFFENCE TYPE"] == "Drug Possession") & (db["YEAR"] == "2020") & (db["variable"].isin(["CASES DISPOSED BEFORE TRIAL DATE WITHDRAWN/ STAYED BEFORE TRIAL", "CASES DISPOSED"]))]
         .pivot(columns="variable", index="REGION", values="value")
         )

drugs_2020["withdrawn_%_2020"] = round(drugs_2020["CASES DISPOSED BEFORE TRIAL DATE WITHDRAWN/ STAYED BEFORE TRIAL"] / drugs_2020["CASES DISPOSED"] * 100, 2)

drugs_2021 = (db
         .loc[(db["OFFENCE TYPE"] == "Drug Possession") & (db["YEAR"] == "2021") & (db["variable"].isin(["CASES DISPOSED BEFORE TRIAL DATE WITHDRAWN/ STAYED BEFORE TRIAL", "CASES DISPOSED"]))]
         .pivot(columns="variable", index="REGION", values="value")
         )

drugs_2021["withdrawn_%_2021"] = round(drugs_2021["CASES DISPOSED BEFORE TRIAL DATE WITHDRAWN/ STAYED BEFORE TRIAL"] / drugs_2021["CASES DISPOSED"] * 100, 2)

drugs_2021.sort_values("withdrawn_%_2021", ascending=False)
```

```python
comparison = drugs_2021.join(drugs_2020, rsuffix="_").loc[:, ["CASES DISPOSED", "withdrawn_%_2020", "withdrawn_%_2021"]]

comparison["change"] = comparison["withdrawn_%_2021"] - comparison["withdrawn_%_2020"]

comparison.sort_values("CASES DISPOSED", ascending=False)
```
