```python
import pandas as pd
```

```python
stations = pd.read_csv('../raw/RAW 2021 ENVIRONMENT CANADA WEATHER STATIONS.csv', encoding="latin-1", header=2)

display(stations.head())
```

```python
manitoba_stations = (stations
            .loc[(stations["Province"] == "MANITOBA") & (stations["Last Year"] == 2021), "Station ID"]
            .to_list()
            )
```

```python
li = []

for station_id in manitoba_stations:
    df = pd.read_csv('https://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=' + str(station_id) + '&Year=2022&timeframe=2')
    df.insert(0, "Station ID", station_id)
    li.append(df)

raw = pd.concat(li, axis=0, ignore_index=True)
```

```python
raw
```

```python
recent = raw.loc[raw["Date/Time"] >= "2022-04-13", :]
recent = recent.pivot(columns="Date/Time", values="Total Snow (cm)", index=["Station ID", "Station Name"]).dropna(axis=1, how="all").sort_values("2022-04-14", ascending=False).dropna(how="all").fillna(0)
recent["April 13-15"] = recent["2022-04-13"] + recent["2022-04-14"] + recent["2022-04-15"]
recent = recent.sort_values("April 13-14", ascending=False)


recent
```
