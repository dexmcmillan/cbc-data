```python
import pandas as pd
from zipfile import ZipFile
import requests
from io import BytesIO
```

```python
files = ZipFile("./data/archive.zip")

print(files)

lis = []

for file in files.namelist():
    print(file)
    file = files.open(file)
    raw = pd.read_csv(file, encoding="latin-1")
    lis.append(raw)
    
data = pd.concat(lis)

data
```

```python
pivot = data.pivot(index="Date", columns="Name", values="Close")
for col, index in pivot.iteritems():
    pivot[col] = pivot[col] / pivot[col].max()

pivot = pivot[pivot.index >= "2021-01-01"]
pivot = pivot[["Bitcoin", "Ethereum"]]
pivot
```
