```python
import pandas as pd
import requests
```

```python
raw = requests.get("https://canopy.cbc.ca/live/elections/prov/ON2022/all/").json()["data"]["ridings"]

raw[0].keys()
```

```python
data = pd.json_normalize(raw, record_path=["parties"], meta=['ridingNumber', 'englishName', 'totalVoters', 'totalVotesReported', 'previousElectedPartyCode', 'pollsReported', "totalPolls"])

data.head(5)
```

```python
ridings = data.drop_duplicates("englishName")

ridings["partyName"] = (ridings["partyCode"]
                        .replace({
                            "LIB": "Liberal Party", 
                            "PC": "Progressive Conservative Party",
                            "GRN": "Green Party",
                            "NDP": "New Democratic Party",
                            "OTH": "Other",
                            "IND": "Independent"
                            })
                        )

ridings["color"] = (ridings["partyCode"]
                    .replace({
                        "LIB": "#ca1b2a",
                        "PC": "#456896",
                        "GRN": "#468240",
                        "NDP": "#db7b14",
                        "OTH": "#89959a",
                        "IND": "#89959a"
                        })
                    )

ridings["polls"] = ridings["pollsReported"]/ridings["totalPolls"]*100

polls_reported = ridings["pollsReported"].mean()/ridings["totalPolls"].mean()*100
turnout = ridings["totalVotesReported"].min() / ridings["totalVoters"].min()

print(polls_reported, turnout)

ridings.to_clipboard()

display(ridings)
```
